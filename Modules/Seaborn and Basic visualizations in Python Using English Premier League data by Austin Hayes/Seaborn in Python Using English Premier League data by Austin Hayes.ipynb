# Barplots

Barplots are different from the first two graphs we have gone over. Instead of analyzing two numerical variables, we are analyzing one numerical variable and a categorical variable. This is a very important graph to use because it allows you to easily compare different categories against one another. 

For example, lets take a look at the average number of goals per position.

If you create a barplot, it will automatically graph the mean amount of the y variable for each category in the x-variable. If you wanted to do other operations such as sum or standard deviation you can use aggregation. 

To do aggregation, we will use .groupby() which allows us to group all of the positions together. Then, we can use ['Goals'] to assign the goals column to that aggregation operation. 

Then, we need to use .mean() in order to tell the .groupby() function what to do with the assigned 'Goals' column. In this case, we are telling it to report the average amount of goals.

Finally we will need to use .reset_index() in order to assign proper integer-based indexing to the aggregated dataframe. When you use aggregation, it can lead to non-numeric indexing which will prevent the use of graphing.  

Below, there will be one example of the aggragated way and the default way. Remember, the default way only works in this case because barplots, by default, report the mean and that is what we want to do here.

Let's start off with the default way:

```python
# Create the barplot
# A barplot is a useful graph that displays the relationship between a numerical variable and a categorical variable.
# In this case, we will use the 'Position' column as the categorical variable and the 'Goals' column as the numerical variable. 
# This will allow us to see the average number of goals scored by each position.
# Use ci=None to remove the confidence interval from the barplot (line that extends above and below the bars)
sns.barplot(data=premier_league_data, x='Position', y='Goals', errorbar=None)

# Let's retitle the y-axis to make it more descriptive
plt.ylabel('Average Goals Scored')
```

Looks like Forwards had the highest average of goals by position during the 2024 - 2025 season!


Now, let's use the aggregated way to find the standard deviation of goals by position:

```python
# Create a new aggregated DataFrame to find the standard deviation amount of goals by position
# This will group the data by 'Position' and the corresponding standard deviation of 'Goals' for each position
# The result is a new DataFrame with the standard deviation of goals for each position
avg_goals_by_position = premier_league_data.groupby('Position')['Goals'].std().reset_index()

# Show the first few rows of the aggregated DataFrame in order to verify the aggregation
avg_goals_by_position.head()
```

```python
# Plot average goals and positions using a barplot
sns.barplot(data = avg_goals_by_position, x = 'Position', y = 'Goals')

# Let's retitle the y-axis to make it more descriptive
plt.ylabel('Standard Deviation Goals Scored')
```

Nice, looks like Forwards not only had the highest average amount of goals, they had the highest standard deviation of goals as well!

NOTE: There is a similar function called countplot that does something similar except it does aggregation automatically and ONLY uses a count/sum on the y-axis. We will review this later in the lesson.

# Histogram


A histogram is a great way of analyzing the distribution of a quantitative variable. Similarly to bar graphs, it is easy to use and interpret. Unlike bar graphs, the x-axis is a continous variable. Data can be grouped into "bins" that accumulate a range of numbers from the variable on the x-axis. The y-axis shows the total frequency of that grouping.

Let's try it out by examining the assists by players in the Premier League!

```python
# Create a histogram to visualize the distribution of player ages
sns.histplot(data=premier_league_data, x='Assists', bins=10)
```

You can also change the number of bins by letting the *histplot* decide the appropriate number of bins via an algorithm.  This can be done by changing the bins=10 argument to bins='auto' when establishing your histogram.

Let's try it!

```python
# Add a trend line to the histogram and let the computer decide the number of bins
sns.histplot(data=premier_league_data, x='Assists', bins='auto')
```

After taking a closer look at the graph, most players have a very small amount of assists. We also notice that the majority of the distribution lies on the left side of the graph. We call this a skewed right distribution. If the histogram was flipped around, it would be called a skewed left distribution. If the data was evenly distributed amongst points, it would be called a uniform distribtion. Finally, if the data appears bell-shaped with the majority of the data being located in the middle and tapering off to the left and right, we would call that a normal distribution.

In this case we can say the following: "According to the Histogram, the distribution of assists in the Premier League during the 2024 - 2025 season is approximately skewed to the right"

# Boxplot

Boxplots are similar to Histograms in that they both explore the distribution of a dataset. However, Boxplots graph the five number summary of a dataset instead. The five number summary includes the following: minimum, maximum, median, 25th and 75th percentile. 


Let's make a Boxplot using the the number of appearances among players with more than 2000 Minutes of game time and analyze the results.

```python
# Create a boxplot using the number of appearances among players
# A boxplot is a useful graph that displays the distribution of a dataset based on a five-number summary
# Filter the data to only include players with more than 2000 minutes of game time
sns.boxplot(data=premier_league_data[premier_league_data['Minutes'] > 2000], x = 'Appearances')
```

The distribution of appearances among players with more than 2000 minutes of game time seems to skwed to the left.

The box in the boxplot represents the Inter Quartile Range (IQR), meaning it contains the 25th percentile, the median and the 75th percentile. The median is the line in the center of the box and the 25th percentile and the 75th percentile are represented on the left and right sides of the box, respectfully. The arms connected to the box, called whiskers, represent data points outside of the IQR. They extend outward until they reach a maximum and minimum point that is one and a half times the IQR. 

When we take a look at the graph, it looks like the median amount of appearances among players last season was around 22! 75th percentile was around 10 and the 25th percentile was around 32!


Although they are not seen in this graph, there are outliers that can be seen outside of the whiskers. 

For example, let's graph assists so that we can see these outliers!

```python
# Graph boxplot of assists
sns.boxplot(data=premier_league_data, x = 'Assists')
```

If you remember from the histogram, this data is heavily skewed to the right so it's median, 25th percentile, and relative minimum are all zero. As you can see, there are outliers outside of the right hand whisker on the boxplot. We can see the true maximum amount of assists a player had this past season was approximately 18.

Pretty neat!

# Countplot

Countplot is a very convenient type of plot in Seaborn because it makes counting categorical very easy. It is very similar to a Barplot except it counts the number of occurances for each category found in the dataset. 


Let's try it out counting the number of players for each football club that appears in the data.

```python
# Make the countplot of players for each team
# A countplot is a bar plot where the x-axis represents categorical data 
# and the y-axis includes a count of occurrences for each category.
sns.countplot(data=premier_league_data, x='Club')
```

Not good; the x labels are all overlapping. Let's fix this using the Matplotlib library, which is what Seaborn was built from.

```python
# Remake the plot
sns.countplot(data=premier_league_data, x='Club')

# Rotate the x labels to prevent overlap
plt.xticks(rotation=90)
```

Much better! We can see that the club with the most amount of players in the dataset is Southampton! The club with the least players is Brighton!

# Violinplot

A violinplot is another useful graph that makes it easy to interpret a datasets distribution. Violin plots use one categorical variable and one numerical variable. The graph illustrutes distribution by using density instead of a boxplot. It is easier explained with an image so let's go ahead and make a violin plot of shots on target during the season among players with more than 22 goals.

```python
# Make the Violinplot of 'Shots On Target' during the season among players with more than 22 goals
sns.violinplot(data = premier_league_data[premier_league_data['Goals'] > 22], x = 'Shots On Target', inner = 'quartile')
```

NOTE: We can add IQR to the violinplot like the one we just made by using a new argument called inner.

Interesting, the Violinplot has an approximately normal distribution. Although the y-axis is not labeled, it is essentially a standardized scale that measures the density / amount of points in the graph at a given amount of shots on target.

Now, lets use a violin plot to compare the distribution(s) of player appearances by position group. We will also filter out the data to ONLY include players with more than 10 goals.

NOTE: We can filter out the rows to only include players that meet any given criteria. In this case, we are interested in seeing the distribution of player appereances by position group, ONLY including players with more than 10 goals. When we filter the data like this, it can drastically change the distribution of our data. If we include a broader amount of players, the IQR would shift towards the direction of the newly added data points.

Now, let's try it out!

```python
# Create a Violinplot to compare Appearances among players with greater than 10 goals by position group
sns.violinplot(data = premier_league_data[premier_league_data['Goals'] > 10], x='Position', y='Appearances', inner='quartile')
```

In this graph, we can conclude that there are were no Goalkeepers nor Defenders that had more than 10 goals last season.

It seems like Midfielders had a left skewed distribution compared to that of Forwards. Interesting!

# Matplotlib Quick Customizations:

As discussed earlier, Seaborn was built from the Matplotlib library. Therefore, some of the Matplotlib functions can be used to help customize the graphs and make them more organized. For example, we used plt.xticks() earlier to rotate the x labels in order to make room for all of the categories on the countplot. 

Here are some of the functions you can use in order to customize your graphs. We will apply these changes to the last Violinplot we made.

```python
# Remake Violinplot from above
sns.violinplot(data = premier_league_data[premier_league_data['Goals'] > 10], x='Position', y='Appearances', inner='quartile')

# label x and y axis
plt.xlabel('Position Group')
plt.ylabel('Number of Appearances')

# Add a title to the plot
plt.title('Shots On Target by Position Group (Players With MIN. 10 Goals)')

# Set the style of the plot to 'whitegrid'
sns.set_style('whitegrid')

# Set the color palette to 'pastel' for a softer look
# This will change the colors of the plot elements to pastel shades
sns.set_palette('pastel')

# Set the font size to 14 for better readability on the y-axis
plt.yticks(fontsize=14)

# Show the plot (ensures that the plot is displayed in the output)
plt.show()
```

Seaborn and Matplotlib are extremely powerful statistical vizualization libraries. There are so many customization options that are available outside of what we have reviewed in this module.

# Scatterplot


A scatterplot is a useful graph that displays the relationship between two numerical variables. Scatterplots are powerful in that they provide a visual for understanding the relationship between two different measures. 

For example, let's say that we want to explore the relationship between the number of minutes played and touches of the ball.

Run the code below:

```python
# Create a scatterplot to visualize the relationship between minutes played and touches made by players
# The x-axis represents the number of minutes played, and the y-axis represents the number of touches made
sns.scatterplot(data = premier_league_data, x='Minutes', y='Touches')
```

As expected, it looks like there is a very strong positive relationship between the number of minutes played and touches of the ball. When a player has more time in game, it makes sense that they've had a higher number of turns with the ball. 


Now let's organize that finding into an interpretation:

"According to the scatterplot, there is a very strong positive relationship between the number of minutes played and touches of the ball among Premier League players during the 2024 - 2025 season."

You can also change the color of the data points with an extra argument called hue. Lets try it out on the scatterplot above in order to see what position each data point is.

```python
# Recreate the scatterplot with a hue based on the player's position
# This will color the points based on the player's position, allowing for a better understanding of how different positions relate to the number of minutes played and touches made
sns.scatterplot(data = premier_league_data, x='Minutes', y='Touches', hue='Position')
```

Very cool!

NOTE: You can also interpret correlation using a coefficient correlation or otherwise called 'r'. 'r' is a number between negative one and positive one that will tell you how your data correlates. Positive numbers mean a positive correlation and negative numbers mean a negative correlation. 



_Here is a range of numbers that can help you interpret each r value:_

0.90 to 1.00 (or -0.90 to -1.00)  =  Very strong	

0.70 to 0.89 (or -0.70 to -0.89)  =  Strong	

0.40 to 0.69 (or -0.40 to -0.69)  =  Moderate

0.10 to 0.39 (or -0.10 to -0.39)  =  Weak

0.00 to 0.09 (or -0.00 to -0.09)  =  Very weak / None

Here is some quick code to help you calculate the r-value using the variables we just used in our scatterplot: 

```python
#Calculate the correlation coefficient (r-value) between minutes played and touches made with the ball
r_value = np.corrcoef(x=premier_league_data['Minutes'],y=premier_league_data['Touches'])

# Print the r-value from the 2D array returned by np.corrcoef
r_value[0,1]  # To access the r value, we can use [0,1] or [1,0] because it is a 2x2 matrix
```

Just like on the scatterplot, we can see that it counts as a very strong correlation using the r-value. Very nice!
