# Scatterplot

A scatterplot is a useful graph that displays the relationship between two numerical variables. Scatterplots are powerful in that they provide a visual for understanding the relationshiop  between two different measures. 

For example, let's say that we want to explore the relationship between succesful crosses and assists.

Run the code below:

```python
# Create a scatterplot to visualize the relationship between successful crosses and assists
sns.scatterplot(data = premier_league_data, x='Successful Crosses', y='Assists')
```

Very Interesting. It seems as though there is a weak positive relationship between the number of successful crosses a player has and their assists. Now let's organize that finding into an interpretation.

"According to the scatterplot, there is a weak positive relationship between the number of successful crosses and assists among Premier League players during the 2024 - 2025 season."

You can also change the color of the data points with an extra argument called hue. Lets try it out on the scatterplot above in order to see what position each data point is.

```python
sns.scatterplot(data = premier_league_data, x='Successful Crosses', y='Assists', hue='Position')
```

Very cool!

NOTE: You can also interpret correlation using a coefficient correlation or otherwise called 'r'. r is a number between negative one and positive one that will tell you how your data correlates. Positive numbers mean a positive correlation and negative numbers mean a negative correlation. 



_Here is a range of numbers that can help you interpret each r value:_

0.90 to 1.00 (or -0.90 to -1.00)  =  Very strong	

0.70 to 0.89 (or -0.70 to -0.89)  =  Strong	

0.40 to 0.69 (or -0.40 to -0.69)  =  Moderate

0.10 to 0.39 (or -0.10 to -0.39)  =  Weak

0.00 to 0.09 (or -0.00 to -0.09)  =  Very weak / None

Here is some quick code to help you calculate the r-value using the variables we just used in our scatterplot: 

```python
#Calculate the correlation coefficient (r-value) between successful crosses and goals scored
r_value = np.corrcoef(premier_league_data['Successful Crosses'],premier_league_data['Assists'])

# Print the r-value from the 2D array returned by np.corrcoef
r_value[0,1]  # To access the r value, we can use [0,1] or [1,0] because it is a 2x2 matrix
```

We can see that it counts as a weak correlation using the r-value. Very nice!

Now let's take a look a lineplots. Lineplots are useful for vizualizing trends among two variables. Unilke scatterplots that just plot all of the datapoints, a lineplot makes a line of best fit that summarizes the behaviour between two numerical variables.

# Barplots

Barplots are different from the first two graphs we have gone over. Instead of analyzing two numerical variables, we are analyzing one numerical variable and a categorical variable. This is a very important graph to use because it allows you to easily compare different categories against one another. 

For example, lets take a look at the number of goals per position.

Before we make the barplot, we need to aggregate the positions and the sum of their goals. This is needed because a bargraph will automatically graph the mean instead of the sum.

To do this we will use .groupby() which allows us to group all of the positions together. Then, we can use ['Goals'] to assignthe goals column to that aggregation operation. 

Then, we need to use .sum() in order to tell the .groupby() function what to do with the assigned 'Goals' column.

Finally we will need to use .reset_index() in order to assign proper integer-based indexing to the aggregated dataframe. When you use aggregation, it can lead to non-numeric indexing which will prevent the use of graphing.  

```python
# Create a new aggregated DataFrame to sum the total goals by position
# This will group the data by 'Position' and sum the 'Goals' for each position
# The result is a new DataFrame with total goals for each position
goals_by_position = premier_league_data.groupby('Position')['Goals'].sum().reset_index()

# Show the first few rows of the aggregated DataFrame in order to verify the aggregation
goals_by_position.head()
```

```python
# Plot total goals and positions using a barplot
sns.barplot(data=goals_by_position, x='Position', y='Goals')
```

Looks like Midfielders scored the most goals during the 2024 - 2025 season!

# Histogram


A histogram is a great way of analyzing the distribution of a quantitative variable. Similarly to bar graphs, it is easy to use and interpret. Unlike bar graphs, the x-axis is a continous variable. Data can be grouped into "bins" that accumulate a range of numbers from the variable on the x-axis. The y-axis shows the total frequency of that grouping.

Let's try it out by examining the assists by players in the Premier League!

```python
# Create a histogram to visualize the distribution of player ages
sns.histplot(data=premier_league_data, x='Assists', bins=10)
```

You can also change the number of bins by letting the computer decide! 

This can be done by changing the bins=10 argument to bins='auto' when establishing your histogram.

Let's try it!

```python
# Add a trend line to the histogram and let the computer decide the number of bins
sns.histplot(data=premier_league_data, x='Assists', bins='auto')
```

After taking a closer look at the graph, most players have a very small amount of assists. We also notice that the majority of the distribution lies on the left side of the graph. We call this a skewed right distribution. If the histogram was flipped around, it would be called a skewed left distribution. If the data was evenly distributed amongst points, it would be called a uniform distribtion. Finally, if the data appears bell-shaped with the majority of the data being located in the middle and tapering off to the left and right, we would call that a normal distribution.

In this case we can say the following: "According to the Histogram, the distribution of assists in the Premier League during the 2024 - 2025 season is approximately skewed to the right"

# Boxplot

Boxplots are similar to Histograms in that they both explore the distribution of a dataset. However, Boxplots graph the five number summary of a dataset instead. The five number summary includes the following: minimum, maximum, median, 25th and 75th percentile. 


Let's make a Boxplot using the the number of minutes among players and analyze the results.

```python
# Make boxplot using the number of minutes players had last season.
sns.boxplot(data=premier_league_data, x = 'Minutes')
```

The distribution of minutes among players seems to be slightly skewed to the right. 

The box in the boxplot represents the Inter Quartile Range (IQR), meaning it contaains the 25th percentile, the median and the 75th percentile. The median is the line in the center of the box and the 25th percentile and the 75th percentile are represented on the left and right sides of the box, respectfully. The arms connected to the box, called whiskers, represent data points outside of the IQR. They extend outward until they reach a maximum and minimum point that is one and a half times the IQR. 

When we take a look at the graph, it looks like the median amount of minutes among players last year was around 1200 minutes! 75th percentile was around 2250 and the 25th percentile was around 300 minutes!


Although they are not seen in this graph, there are outliers that can be seen outside of the whiskers. Let's graph Assists so that we can see these outliers!

```python
# Graph boxplot of assists
sns.boxplot(data=premier_league_data, x = 'Assists')
```

If you remember from the histogram, this data is heavily skewed to the right so it's median, 25th percentile, and relative minimum are all zero. As you can see, there are outliers outside of the right hand whisker on the boxplot. We can see the true maximum amount of assists a player had this past season was approximately 18.

Pretty neat!

# Countplot

Countplot is a very convenient type of plot in Seaborn because it makes counting categorical very easy. It is very similar to a Barplot except it counts the number of occurances for each category found in the dataset. 


Let's try it out counting the number of players for each football club that appears in the data.

```python
# Make the countplot of players for each team
# A countplot is a bar plot where the x-axis represents categorical data 
# and the y-axis includes a count of occurrences for each category.
sns.countplot(data=premier_league_data, x='Club')
```

Not good; the x labels are all overlapping. Let's fix this using the Matplotlib library, which is what Seaborn was built from.

```python
# Remake the plot
sns.countplot(data=premier_league_data, x='Club')

# Rotate the x labels to prevent overlap
plt.xticks(rotation=90)
```

Much better! We can see that the club with the most amount of players in the dataset is Southampton! The club with the least players is Brighton and Hove!

# Violin plot

A violinplot is another useful graph that makes it easy to interpret a datasets distribution. Violin plots use one categorical variable and one numerical variable. The graph illustrutes distribution by using density instead of a boxplot. It is easier explained with an image so let's go ahead and make a violin plot of Offside penalties during the season.

```python
# Make the violin plot of Offside penalties during the season
sns.violinplot(data = premier_league_data, x='Offsides', inner='quartile')
```

NOTE: We can add IQR to the violinplot like the one we just made by using a new argument called inner.

Interesting, the violinplot is skewed right. Although the y-axis is not labeled, it is essentially a standardized scale that measures the density / amount of points in the graph at a given amount of offside penalties.

Now, lets use a violin plot to compare offside penalties among position groups.

Let's get started!

Now, that we have aggregated them lets take a look

```python
sns.violinplot(data = premier_league_data, x='Position', y='Offsides', inner='quartile')
```

Goalkeepers cannot be given an offside penalty, hence why it has no distribution. 

Just like in the previous violinplot, each position groups offside penalties are skewed to the right. However, defenders (DEF) seem to be more clustered together with a higher density.

# Matplotlib Quick Customizations:

As discussed earlier, Seaborn was built from the Matplotlib library. Therefore, some of the Matplotlib functions can be used to help customize the graphs and make them more organized. For example, we used plt.xticks() earlier to rotate the x labels in order to make room for all of the categories on the countplot. 

Here are some of the functions you can use in order to customize your graphs. We will apply these changes to the violinplot graph.

```python
sns.violinplot(data = premier_league_data, x='Position', y = 'Offsides', inner='quartile')

# label x and y axes
plt.xlabel('Position Group')
plt.ylabel('Offside Pentalties')

# Add a title to the plot
plt.title('Offside Penalties by Position Group')

# Set the style of the plot to 'whitegrid'
sns.set_style('whitegrid') 

# Set the color palette to 'pastel' for a softer look
# This will change the colors of the plot elements to pastel shades
sns.set_palette('pastel') 

# Set the font size to 14 for better readability on the y-axis
plt.yticks(fontsize=14)    

# Show the plot
plt.show()
```
