```python
# Import necessary libraries

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

### 1. Import the La Liga dataset and create a scatterplot. Use 'Successful tackles' on the x-axis and 'Successful duels' on the y-axis in order to explore the relationship between the two.

```python
# Import the La Liga dataset
la_liga_data = pd.read_csv('https://raw.githubusercontent.com/schuckers/Charlotte_SCORE_Summer25/refs/heads/main/Data%20for%20Modules/Data%20for%20Seaborn%20and%20Basic%20visualizations%20in%20Python%20Using%20English%20Premier%20League%20data%20by%20Austin%20Hayes/laliga_player_stats_english.csv')
```

```python
la_liga_data.columns
```

```python
# Create a scatterplot with 'Successful tackles' on the x-axis and 'Goals Scored' on the y-axis
sns.scatterplot(data=la_liga_data, x='Successful tackles', y='Successful duels')

# Set the title for the scatterplot 
plt.title('Successful Tackles vs Goals Scored')
```

### 2. Using the scatterplot you created in the first problem, is there any correlation between 'Successful tackles' and 'Goals scored'? Write out a short interpretation using 1-2 sentences.

```python
np.corrcoef(x=la_liga_data['Successful tackles'], y=la_liga_data['Successful duels'])
```

According to the scatterplot, there is a strong positive correlation between the number of successful tackles and the number of successful duels a player has.

### 3. Make a barplot that allows you to make compare the total amoung of red cards each team has. What team has the most red cards?

Hint: Use aggregation. You will also need to use a certain function to flip the x labels

```python
# Aggregate the total number of red cards for each team
# This will create a new DataFrame with the total red cards for each team
aggregated_team_rc_sum = la_liga_data.groupby('Team')['Red Cards'].sum().reset_index()

# Verify the aggregation
aggregated_team_rc_sum
```

```python
# Make the barplot
sns.barplot(data=aggregated_team_rc_sum, x='Team', y='Red Cards')

# Rotate the x labels because they would overlap otherwise
plt.xticks(rotation=90)
```

Rayo Vallecano has the most red cards

### 5. Make a histogram and a boxplot to explore the distribution of offside penalties. What does the distribution look like? Is it skewed left or right? Or is it a normal distribution?

```python
# Make a boxplot to explore the distribution of offside penalties
sns.boxplot(data=la_liga_data, x='Offsides')
```

```python
# Make a histogram to explore the distribution of offside penalties
sns.histplot(data=la_liga_data, x='Offsides', bins='auto')
```

It is EXTREMELY skewed to the right.
