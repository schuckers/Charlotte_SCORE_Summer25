```python
# Import needed libraries 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv("https://raw.githubusercontent.com/schuckers/Charlotte_SCORE_Summer25/refs/heads/main/Data%20for%20Modules/Data%20for%20Medal%20Distribution%20in%20Olympic%20Wrestling%20by%20Izaan%20Khudadad/olympic-freestylewrestling-results-1894-2020-v1.csv")

# See the first 5 rows
df.info()
```

1. Gold Medal Counts by Country

```python
# Step 2: Check unique values in medal-related column
print(df['medal_type'].unique())
# Filter for gold medals (adjust if labels differ)
gold_medals = df[df['medal_type'].str.contains("Gold", case=False, na=False)]

# Count gold medals per country
gold_counts = gold_medals['country_name'].value_counts()
```

- Use the wrestling dataset to calculate how many gold medals each country has won.
- Make a bar chart showing which countries have the most gold medals.

```python
# Plot bar chart of top 15 countries by gold medals
plt.figure(figsize=(10,6))
gold_counts.head(15).plot(kind='bar', color='gold', edgecolor='black')
plt.title("Top Countries by Gold Medals (Freestyle Wrestling)")
plt.xlabel("Country")
plt.ylabel("Gold Medal Count")
plt.xticks(rotation=45, ha='right')
plt.show()
```

- Calculate the Gini coefficient for gold medal distribution.

```python
# Step 4: Gini Coefficient
values = gold_counts.values
sorted_vals = np.sort(values)
cum_vals = np.cumsum(sorted_vals)
n = len(values)
gini = (n + 1 - 2 * (cum_vals / cum_vals[-1]).sum() / n)

print("Gini Coefficient for Gold Medals:", round(gini, 3))
```
