# Explaination of Statisitics

Before we begin the technical aspect of this learning module, let's go over some key statistics and what they mean. 

_Field Goal_: Any type of basket made during live basketball play. Any type of basket made outside of free throws is considered a field goal. Dunks, layups, mid-range shots and three-pointers are all examples of a field goal.

_2-point shot_: A 2 point shot is any type of field goal that counts for 2 points. These shots are made from inside the 3-point line, which is the the large line painted around a basketball net on one side of a court. 2-point shots includes dunks, layups and midrange shots.

_3-point shot_: A 3 point shot is the reverse of a 2-point shot. It is any shot made outside of the 3-point line. If a player steps over the line during the shot, it will count as a 2-point shot. 

_Rebound_: A rebound is when a player shoots the ball, misses the shot and the same or different player catches the ball after it falls down from the rim. The player who made the rebound must maintain full control of the ball and land on the ground with possesion. 

_Assist_: An assist is made when one player passes the ball to their teammate and that teammate makes a field goal. However, the pass must directly contribute to the field goal that was made. For example, the general rules state that an assist only counts if the teammate who recieved the ball via a pass, makes the field goal in 2 dribbles or less. 

_Free throw_: A free throw is a free, unguarded shot (or 2) that was award to a player via a foul commited by the other team. Usually, free throws are awarded off of shooting fouls which occur when a player attempts to score a basket but a defender on the opposing team makes illegal contact, such as a slap on the wrist while a player is making a shot attempt. 

_Block_: A block is made when a defender swipes an offensive players shot out of the air without making illegal contact with the player who attempted the shot. 

_Steal_: A steal occurs when a defensive player cleanly swipes the ball out of an offensive players hands or when a defensive player intercepts a ball off a pass between two offensive players. 

_Turnover_: A turnover occurs when there is a steal by the defense, a travel violation by the offense (more then two steps without dribbling the ball), an offensive player stepping out of bounds, etc.

We need to scale the data in order to even the playing field among variables for model-making. This prevents a model from becoming too biased towards features or variables with larger numerical ranges. Scaling essentially standardizes the data so that all data points are rescaled so that their means are 0 and the standard deviation is 1. Although the data is rescaled, their data ranges are not distorted. The formula is the following: (x - mean) / Standard Deviation. 

Note on fit_transform vs ()transform().

fit_transform() scales the data as detailed above while also calculating the unique parameters of the data. These parameters will be used to scale the test set, which creates new and unique results for the model to make predictions on. This creates less biased and mroe well-rounded results. 

transform() this scales the data as detailed above, but without the fit portion of the function above. This data will be rescaled using the training set's parameters. 

We will use fit_transform() on the training data and transform() on the testing data.

# K Nearest Neigbors (KNN)



K Nearest Neighbors is a type of model that decides where to classify a data point based on the distance from previously clssified data points that the model was trained on. It is essentially deciding where to classify a data point based on the closest points or "neighbors". You can control the number of neighbors that a data point will be classified with by defining K. If you make the K number of neighbors too high, it may make the model too generalized and cause it to have low accuracy. Likewise, if you make K too small, it will be too niche and become inaccurate as well. 

Due to KNN being distance based, you will need to scale the predictor variables in order to get the best results. Scaling data points essentially levels the playing field among different measures in order to prevent the model from becoming too biased towards predictor variables with larger measures. For example, 'FGA', or 'Field Goals Attempted Per Game' will always have a larger range of numbers than 'Offensive Rebounds Per Game'

Let's test it out.

IMPORTANT RULE: For good practice, you should ONLY use odd numbers for K. This is done to prevent a tie during the prediction process. For example, if you used K=4 and there happens to be a 2-2 vote among points, it will likely result in an inaccuracte outcome. 

We will need to use standard scaler to scale the data. Let's go ahead and make the KNN model. Let's start off with K = 5 Neighbors.
